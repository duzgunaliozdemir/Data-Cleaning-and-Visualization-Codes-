```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```
vis=pd.read_excel("datacleaned.xlsx")
```

**1) How Much Are Different Races Paid Monthly?**


```

sns.set(rc={'figure.figsize':(15,7.5)})

tips = sns.load_dataset("tips")
sns.set_style("white")

sns.violinplot(data=vis, y="Race", x="Monthly Income", hue="Gender",
               split=True, inner="quart", linewidth=1,
               palette={"Female": "#ff9d9a", "Male": "#72a3ff"}, figsize=(20,20))
sns.despine(left=True)

plt.title("Monthly Income Distribution by Gender for Different Races", fontsize=18,fontweight="bold")
plt.xlabel("Monthly Income",fontsize=14 ,fontweight="bold")
plt.ylabel("Race",fontsize=14 , fontweight="bold")

plt.show()
```

**2) Which Age Group Works in Different Positions?**

```
sns.set(rc={'figure.figsize':(15,13)})
sns.set_style("white")
sns.boxplot(data=vis, x="Position", y="Age", hue="Gender", palette=["#ff9d9a","#72a3ff"])
plt.title("Age Distribution in Different Positions", fontsize=18, fontweight="bold")
plt.xlabel("Position", fontsize=13, fontweight="bold")
plt.ylabel("Age",fontsize=13, fontweight="bold")
plt.show()
```

**3) How Is the Perform Evaluation of the Genders Distributed by Positions?**

```
sns.set(rc={'figure.figsize':(15,7.5)})
sns.set_style("white")
sns.barplot(data=vis, x="Position",y= "Perform Evaluation", hue="Gender", palette=["#ff9d9a","#72a3ff"], ci= None )
plt.title("Performance of Workers in Different Positions", fontsize=18, fontweight="bold" )
plt.xlabel("Position",fontsize=13, fontweight="bold")
plt.ylabel("Peform Evaluation",fontsize=13, fontweight="bold")
plt.show()
```
